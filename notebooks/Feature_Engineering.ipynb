```python
def get_team_stats(team , previous_matches):
    
    team_matches = previous_matches[(previous_matches["team1"] == team) | (previous_matches["team2"] == team)]
    
    matches_played = len(team_matches)

    wins = (team_matches["winner"] == team).sum()

    if matches_played == 0:
        win_rate = 0
    else:
        win_rate = wins / matches_played

    last5 = team_matches.tail(5)

    team_form = (last5["winner"] == team).sum()

    return {
    "matches_played": matches_played,
    "wins": wins,
    "win_rate": win_rate,
    "form_5": team_form
}
```

```python
team1_stats = get_team_stats(team1, previous_matches)

team1_winrate = team1_stats["win_rate"]
team1_form = team1_stats["form_5"]

print(team1_stats)
print(f"Team 1 Win Rate: {team1_winrate}")
print(f"Team 1 Form: {team1_form}")
```

```python
team2_stats = get_team_stats(team2, previous_matches)

team2_winrate = team2_stats["win_rate"]
team2_form = team2_stats["form_5"]

print(team2_stats)
print(f"Team 2 Win Rate: {team2_winrate}")
print(f"Team 2 Form: {team2_form}")
print
```

```python
team1_matches_played_list = []
team1_win_rate_list = []
team1_form_5_list = []

team2_matches_played_list = []
team2_win_rate_list = []
team2_form_5_list = []

for row in range(len(df)):
    current_match = df.iloc[row]

    team1 = current_match["team1"]
    team2 = current_match["team2"]

    previous_matches = df.iloc[:row]

    team1_stats = get_team_stats(team1, previous_matches)

    team2_stats = get_team_stats(team2, previous_matches)

    team1_matches_played_list.append(
    team1_stats["matches_played"]
    )

    team1_win_rate_list.append(
    team1_stats["win_rate"]
    )

    team1_form_5_list.append(
    team1_stats["form_5"]
    )

    team2_matches_played_list.append(
    team2_stats["matches_played"]
    )

    team2_win_rate_list.append(
    team2_stats["win_rate"]
    )

    team2_form_5_list.append(
    team2_stats["form_5"]
    )
# print(len(team1_matches_played_list))
# print(len(team1_win_rate_list))
# print(len(team1_form_5_list))

# print(len(df))
df["team1_matches_played"] = team1_matches_played_list
df["team1_win_rate"] = team1_win_rate_list
df["team1_form_5"] = team1_form_5_list

df["team2_matches_played"] = team2_matches_played_list
df["team2_win_rate"] = team2_win_rate_list
df["team2_form_5"] = team2_form_5_list

df.head(5)

```

```python
def h2h_match_stats(team1, team2, previous_matches):

    h2h_match = previous_matches[
     ((previous_matches["team1"] == team1) & (previous_matches["team2"] == team2)) |
    ((previous_matches["team1"] == team2) & (previous_matches["team2"] == team1))
    ]
    
    h2h_t1_wins = (h2h_match["winner"] == team1).sum()
    h2h_t2_wins = (h2h_match["winner"] == team2).sum()
    return {
        "team1_h2h_wins": h2h_t1_wins,
        "team2_h2h_wins": h2h_t2_wins
    }


row = 100

current_match = df.iloc[row]

team1 = current_match["team1"]
team2 = current_match["team2"]

previous_matches = df.iloc[:row]

h2h_stats = h2h_match_stats(
    team1,
    team2,
    previous_matches
)

print(h2h_stats)
```

```python
def get_venue_stats(team, venue, previous_matches):
    

    team_venue_matches = previous_matches[
    (
        (previous_matches["team1"] == team) |
        (previous_matches["team2"] == team)
    )
    &
    (previous_matches["venue"] == venue)
    ]
    
    matches_played = len(team_venue_matches)

    wins = (team_venue_matches["winner"] == team).sum()

    venue_win_rate = wins / matches_played


    if matches_played == 0:
        return{
        "venue_matches": 0,
        "venue_win_rate": 0
    }


    return{
        "venue_matches" : matches_played,
        "venue_win_rate" : venue_win_rate

    }


venue_stats = get_venue_stats(
    team1,
    venue,
    previous_matches
)

print(venue_stats)
```

```python
team1_venue_matches_list = []
team1_venue_win_rate_list = []

team2_venue_matches_list = []
team2_venue_win_rate_list = []

for row in range(len(df)):
    current_match = df.iloc[row]

    previous_matches = df.iloc[:row]

    team1 = current_match["team1"]
    team2 = current_match["team2"]
    venue = current_match["venue"]

    team1_venue_stats = get_venue_stats(
    team1,
    venue,
    previous_matches
    )

    team2_venue_stats = get_venue_stats(
    team2,
    venue,
    previous_matches
    )

    team1_venue_matches_list.append(
    team1_venue_stats["venue_matches"]
)

    team1_venue_win_rate_list.append(
    team1_venue_stats["venue_win_rate"]
)

    team2_venue_matches_list.append(
    team2_venue_stats["venue_matches"]
)

    team2_venue_win_rate_list.append(
    team2_venue_stats["venue_win_rate"]
)

print(len(team1_venue_matches_list))
print(len(team1_venue_win_rate_list))
print(len(team2_venue_matches_list))
print(len(team2_venue_win_rate_list))

print(len(df))
```
