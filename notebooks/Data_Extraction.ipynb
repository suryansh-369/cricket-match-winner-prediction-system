```python
import os
import json

folder = "t20s_json"

matches = []

for file in os.listdir(folder):

    if file.endswith(".json"):

        with open(os.path.join(folder, file), "r", encoding="utf-8") as f:

            data = json.load(f)

            match = {
                "team1": data["info"]["teams"][0],
                "team2": data["info"]["teams"][1],
                "venue": data["info"]["venue"],
                "toss_winner": data["info"]["toss"]["winner"],
                "toss_decision": data["info"]["toss"]["decision"],
                "winner": data["info"]["outcome"]["winner"] if "winner" in data["info"]["outcome"] else None,
                "date": data["info"]["dates"][0]
            }

            matches.append(match)

print(len(matches))
```

```python
df = pd.DataFrame(matches)
df.shape


```

```python
df = df.dropna(subset=["winner"])
df.shape
df.info()
df.isnull().sum()
```

```python
df.to_csv("matches_clean.csv", index=False)
```
